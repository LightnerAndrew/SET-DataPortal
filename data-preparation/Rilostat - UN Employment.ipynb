# Employment Data - Rilostat

This R notebook accesses the ILOstat database using the Rilostat R package.  We then export the data as a CSV file to be included within the full database.  

#### Preamble

```r
library('Rilostat')
```

### Getting started

This links to a great tutorial on how to use the package. 

```r
ilostat('GettingStarted')
```

### Search for the dataset of choice

```r
toc <- get_ilostat_toc(search = 'EMP_2EMP_SEX_ECO_NB')
```

After searching through the results below, we choose 'EMP_2EMP_SEX_ECO_NB_A' - Employment by sex and economic activity -- ILO modelled estimates, Nov. 2017 (thousands)

```r
toc
```

```r
UN_emp <- get_ilostat(id = 'EMP_2EMP_SEX_ECO_NB_A', segment = 'indicator') 
```

```r
UN_emp
```

## Export Data 

```r
write.csv(UN_emp, "./Datasets/UN_employment.csv")
```
