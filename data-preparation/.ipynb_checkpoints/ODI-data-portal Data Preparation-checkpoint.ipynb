### 1.2.1.1 Initial Access - UNdata Webpage

```python
#Load in some libraries to handle the web page requests and the web page parsing...
import requests

#You may need to install BeautifulSoup
#!pip3 install beautifulsoup4
from bs4 import BeautifulSoup

#Note - I'm in Python3
from urllib.parse import parse_qs

#The scraper will be limited to just the first results page...
def searchUNdata(q):
    ''' Run a search on the UN data website and scrape the results '''
    
    params={'q':q}
    url='http://data.un.org/Search.aspx'
    

    response = requests.get(url,params=params)

    soup=BeautifulSoup(response.content, "lxml")

    results={}

    #Get the list of results
    searchresults=soup.findAll('div',{'class':'Result'})
    
    #For each result, parse out the name of the dataset, the datamart ID and the data filter ID
    for result in searchresults:
        h2=result.find('h2')
        #We can find everything we need in the <a> tag...
        a=h2.find('a')
        p=parse_qs(a.attrs['href'])
        results[a.text]=(p['d'][0],p['f'][0])

    return results

```

### Use the searchUNdata function to look for the necessary employment and labour productivity varaibles

This code scrapes the UN webpage and returns the database options given a search topic.  We will explore the code with 'employment'. 

```python
# A couple of helper functions to let us display the results

results=searchUNdata('gross value added')

def printResults(results):
    ''' Nicely print the search results '''
    
    for result in results.keys():
        print(result)


def unDataSearch(q):
    ''' Simple function to take a search phrase, run the search on the UN data site, and print and return the results. '''
    
    results=searchUNdata(q)
    printResults(results)
    return results

printResults(results)
```

### Select Correct Database and Place within Search Funtion 

Copy 'Gross Value Added by Kind of Economic Activity at constant (2010) prices - US dollars' and place within the search function. 

#### Create Search Function 

```python
#Note - I'm in Python3
from io import BytesIO

import zipfile
import pandas as pd

def getUNdata(undataSearchResults,dataset):
    ''' Download a named dataset from the UN Data website and load it into a pandas dataframe '''

    datamartID,seriesRowID=undataSearchResults[dataset]
    
    url='http://data.un.org/Handlers/DownloadHandler.ashx?DataFilter='+seriesRowID+'&DataMartId='+datamartID+'&Format=csv'

    r = requests.get(url)
    
    
    s=BytesIO(r.content)
    z = zipfile.ZipFile(s)
    
    #Show the files in the zip file
    #z.namelist()
    
    #Let's assume we just get one file per zip...
    #Drop any all blank columns
    df=pd.read_csv( BytesIO( z.read( z.namelist()[0] ) )).dropna(axis=1,how='all')
    return df
```

Oddly, the total GVA variable does not equal the sum of its parts.  Thus we will replace the varaible with a new sum. Additionally, we will save the difference bewteen the two values in a new dataset which will be exported as a csv.

```python
n = 0
for i in sector_vars: 
    # in percentage terms
    UN_gva['Proportion of GVA: '+i] = UN_gva[i]/UN_gva['GVA: Total (ODI)']*100
    UN_gva['GVA: '+i] = UN_gva[i]
    UN_gva.drop(i, axis=1, inplace=True)
    sector_vars[n] = 'GVA: '+i
    n+=1

for i in (list(sector_vars)+['GVA: Total (ODI)', 'Total Value Added']):
    UN_gva[i] = UN_gva[i]/1000
```

# 2. Generate Structural Transformation Measures 

This section generates the structural transformation variables theorized as important to measure structural transformation in developing states.  These measures include: 

- 1) Labour productivity and Labour Productivity Growth
- 2) Relative Labour Productivity
- 3) Within/Between Sector Labour Productivity Growth (Type 1 and 2) 


### 2.1 Labour Productivity and Annualised Labour Productivity Growth

#### 2.1.2 Labour Productivity

Our measure of labour productivity growth is based 
on Haltiwanger (1997), and Foster, Haltiwanger and 
Krizan (2001), who use this decomposition to explore the 
relationship and dynamics between firm-level and aggregate 
productivity growth using firm-level panel data with high 
sample entry and attrition rates. 

Assuming that total labour 
productivity is given by:

#### (1) $P_{t} =\sum_{i=1}^{n} \theta_{i,t}P_{i,t}  $

where $P_{t}$ is total labour productivity in year $t$, $\theta_{i,t}$ denotes the proportion of total labour employed in sector $i$ at time $t$, and $P_{i,t}$ denotes labour productivity in sector $i$ at time $t$; where $i=1,...,9$. 

##### Basic cleaning

```python
# Rename Total Variabe
initial.rename(columns = {'Employment in All (Thousands): Male and Female': 'Employment Total (Thousands): Male and Female', 'Proportion of Employment (%): Employment in All (Thousands): Male and Female': 'Proportion of Employment (%): Employment in Total (Thousands): Male and Female'}, inplace=True)

# Generate list of variables for employment and gva data by thousands. 
filter_col_employ = [col for col in initial if (col.startswith('Employment')) & (col.endswith('Male and Female'))]
filter_col_employprop = [col for col in initial if (col.startswith('Proportion of Employment')) & (col.endswith('Male and Female'))]
filter_col_gva = [col for col in initial if col.startswith('GVA: ')]


# Reorder list to match (proportion)
total = filter_col_employprop[1]
filter_col_employprop.remove('Proportion of Employment (%): Employment in Total (Thousands): Male and Female')
filter_col_employprop = [total] + filter_col_employprop


# Reorder list to match (employment)
total = filter_col_employ[1]
filter_col_employ.remove('Employment Total (Thousands): Male and Female')
filter_col_employ = [total] + filter_col_employ

# Reorder list to match (GVA)
filter_col_gva = [filter_col_gva[0]]+ sorted(filter_col_gva[1:])
```

##### Create Labour Productivity Measure

```python
LPvars= []
# Although we have employment data for 2017, GVA data only exists for 2016. Thus, LP vars only exist until 2016. 
for i in range(0, len(filter_col_gva)): 
    # LP = GVA/Employment for each sector
    initial['Labour Productivity: '+filter_col_gva[i][4:]] = initial[filter_col_gva[i]]/initial[filter_col_employ[i]]
    LPvars.append('Labour Productivity: '+filter_col_gva[i][4:])
```

Ensure equation (1) holds

```python
checkvars = []

# Check proper calclutation by adding the subcomponents together = should add to Total Var. 
check = initial[LPvars]
check.head()

for i in range(1,len(filter_col_employprop)):
    check[LPvars[i]+'prop'] = initial.loc[:, lambda df: LPvars[i]]* (initial.loc[:, lambda df: filter_col_employprop[i]])/100
    checkvars.append(LPvars[i]+'prop')

check = check.loc[:, lambda df: checkvars].sum(axis=1)
check = check.to_frame()

check['LP'] = initial['Labour Productivity:  Total (ODI)']
check.head()
```

### 2.1.2 Annualised Change in Labour Productivity and Employment Share

This indicator presents a simple annualised growth variable from year to year.  Note that for all annualised growth measures, a year observation refers tot change from the year prior to that year. For example, 1995 refers to the growth from 1994-1995. 

```python
initial.reset_index(inplace=True)

# set index to unique observations
initial.sort_values(['countryname', 'year'], inplace=True)
initial.set_index(['countryname', 'year'], inplace=True)

# Generate list of WB vars for change
WBchange = ['GNI (constant 2010 US$)',  'Foreign direct investment, net inflows (% of GDP)', 'Foreign direct investment, net inflows (BoP, current US$)', 'Merchandise exports (current US$)',]


# Generate list of vars for change function
change = LPvars+filter_col_employprop[1:]+WBchange
```

## 2.3 Wtihin and Between Labour Productivity Growth

"Through the exmaination of the relationship and dynamics between firm-level and aggregate level productivity growth, Haltiwanger (1997), and Foster, Haltiwanger and Krizan (2001) show that change in labour productivity can be dissagregated between *within-sector* and *between-sector* productivity changes. 

If equation (1) holds, then the change in total labour productivity between  $t$ and $t-k$ ($\Delta P$) can be written as: 

#### (2) $\Delta P = \sum_{i=1}^{n} \theta_{i,t-k}\Delta P_{i,t} +  \sum_{i=1}^{n} \theta_{i,t}\Delta P_{i,t-k} + \sum_{i=1}^{n} \theta_{i,t}\Delta P_{i,t} ...$

where the first term on the right hand side (RHS) captures 
within
-sector productivity changes; the second term on the 
RHS captures 
between
-sector productivity changes; and the 
third term on the RHS captures cross-sector productivity 
changes. A negative cross term would indicate either that 
labour (employment shares) is moving to sectors where 
productivity fell, or that labour (employment shares) is 
moving out of sectors where productivity increased. A 
positive 
cross
 term would tell us that employment shares 
are growing in sectors where productivity increased, or that 
employment shares are falling in sectors with decreasing 
productivity. Productivity is expressed in real term, using 
real value addition data. 

Combining the second and third terms in the second 
equation (above), we can express labour productivity 
growth using the following decomposition: 

#### (3) $\Delta P = \sum \theta_{i,t-k}\Delta P_{i,t} + \sum_{i=1}^{n} \theta_{i,t}\Delta P_{i,t} ...$

where $P_{t}$ and $P_{i,t}$ refer to economy wide and sectoral labour productivity levels, respectivity, and $\theta_{i,t}$ is the share of employment in sector $i$. The $\Delta$ denotes the change in productivity or employment shares between $t-k$ and $t$. 


The first term in the decomposition is the weighted 
sum of productivity growth within individual sectors, 
where the weights are the employment share of each sector 
at the beginning of the time period. Following McMillan 
and Rodrik (2011), we call this the ‘within’ component 
of productivity growth. The second term captures the 
productivity effect of labour reallocations across different 
sectors. It is essentially the inner product of productivity 
levels (at the end of the time period) with the change 
in employment shares across sectors. When changes 
in employment shares are positively correlated with 
productivity levels, this term will be positive. Structural 
change will increase economy-wide productivity growth. 
Also following McMillan and Rodrik (2011), we call this 
second term the ‘structural change’ term.

(Page 16 - SET Approach Paper)."

#### List of Necessary Variables

```python
# Annualised Change in Labour productivity
filter_col_gLP = [col for col in initial if col.startswith('Annualised Change in Labour')]

# Annualised Change in Employment Share
filter_col_gemploy = [col for col in initial if col.startswith('Annualised Change in Prop')]

# Proportion of Employment
employprp = filter_col_employprop 

# Relative LP 
RelLP = rel
```

```python
# Generate a zero variable for change in employment total
initial['Total Change in Employment'] = 0
filter_col_gemploy = ['Total Change in Employment'] + filter_col_gemploy
```

### Method 1 - Within/Between Growth

There are two forms to decompose labour productivity into their within and between components. The first uses the following equation. 

Note: We leave the proportion variable scaled by 100 to make the variables in percentage terms. 

```python
# list of varsw
within1 = []
between1 = []
total1 = []

for i in range(0, len(filter_col_gLP)): 
    # Annualised growth in labour productivity (absolute) x employment share in earlier year
    initial['Within (1) '+filter_col_gLP[i]] =  initial[filter_col_gLP[i]] * initial[employprp[i]].shift(1)
    
    # Annualised growth in labour productivity (absolute) minus within sector change
    initial['Betewen (1) '+filter_col_gLP[i]] = initial[filter_col_gLP[i]] - initial['Within (1) '+filter_col_gLP[i]]
    
    # Sum of within and between
    initial['Total (1) '+filter_col_gLP[i]] = initial['Within (1) '+filter_col_gLP[i]] +initial['Betewen (1) '+filter_col_gLP[i]]
    
    # Vars to list 
    within1.append('Within (1) '+filter_col_gLP[i])
    between1.append('Betewen (2) '+filter_col_gLP[i])
    total1.append('Total (2) '+filter_col_gLP[i])
```

### Method 2 - Within/Between Growth
 
The second method weights the within and between measures by the relative productivity of that sector.  

```python
# list of vars
within2 = []
between2 = []
total2 = []
for i in range(0, len(filter_col_gLP)): 
    # Annualised growth in labour productivity (absolute) x employment share in earlier year x relative productivity level in earlier year
    initial['Within (2) '+filter_col_gLP[i]] =  initial[filter_col_gLP[i]] * initial[employprp[i]].shift(1)*initial[RelLP[i]].shift(1)
    
    #Annualised change in employment share x relative productivity level in earlier year x employment share in earlier year x (1 + annualised growth in labour productivity (absolute))
    initial['Betewen (2) '+filter_col_gLP[i]] = initial[filter_col_gemploy[i]]*initial[RelLP[i]].shift(1)*initial[employprp[i]].shift(1)*(1+initial[filter_col_gLP[i]])
    
    # Sum of within and between
    initial['Total (2) '+filter_col_gLP[i]] = initial['Within (2) '+filter_col_gLP[i]] +initial['Betewen (2) '+filter_col_gLP[i]]
    
    within2.append('Within (2) '+filter_col_gLP[i])
    between2.append('Betewen (2) '+filter_col_gLP[i])
    total2.append('Total (2) '+filter_col_gLP[i])
```

## 4.2 Cleaning Procedures

There are three main decision to make when generating country aggregates; how to deal wtih large outliers, missing values, and country weights.  Here we address those issues. 

#### A) Dropping Outliers 

We drop all observations beyond 4 standard deviations away from the mean of a variable for a given indicator. Outside of a few variables such as number of battle deaths, this adjustment has only modest effects on the aggregates. 

#### B) Missing Values

We generate a aggregate value only if over 65 percent of the country observation which make up the aggregate are present. This ensures the estimated aggreagte bares some ressemblance to the true aggregate. 

#### C) Country Weights

Theory suggests that if the aim of the study is policy related, then weights play a smaller role when creating aggregates.  This is because we think of national level policy as the observation; and thus there is little use in weighting the country-aggregates.

## Generate Tax CSV file 

The tax varialbes are not particularly clean.  In particular, the total sums of the components to the tax variables do not add to one - similar to the export data. 

Moreover, several countries do not have records for resource dissagregated data.  As such we generate two sets of proportions: those including resource income and those whose resource income is included in the other revenue categories.  

We perform a similar operation as the export variables where a new aggregate is generated by summing the parts which (should) add to 100 percent.  Then we generate new proportions in reference to this new sum. 

#### Load full tax dataset f om the International Centre for Tax and Development


```python
path = '/media/lightner/TOSHIBA EXT/ODI/Report/Data/'
df_tax = pd.read_stata(join(dirname(path), 'ICTD_GRD_data.dta'))

# Only choose observations whose total revenue is less than 100 percent. 
df_tax = df_tax[df_tax['rev_inc_sc']<100]

#We drop all missing values since is any missing values exist we cannot calculate an accurate sum. 
tax_vars = ['identifier', 'rev_inc_sc', 'tot_resource_rev','direct_inc_sc',  'income_tax_share', 'direct_inc_sc_ex_resource_rev', 'id', 'year', 'tot_nresource_rev_inc_sc', 'countrycode', 'tax_g_s', 'tax_int_trade_trans']

tax_df = df_tax[tax_vars]
```

### Generate other variable for both types of countries 

Because the addition do not add to 100, we generate a 'other' category to sum to 100. 

As stated, we generate visuals for 'including resources disaggregation' and non-resource dissagregation.  Thus, we generate two other variables. 

### Generate new percentage variables including the other category.

```python
tax_meas = ['tot_resource_rev', 'direct_inc_sc', 'direct_inc_sc_ex_resource_rev', 'tot_nresource_rev_inc_sc','tax_g_s', 'tax_int_trade_trans']

# Generate the 'as a percentage of revenue variables' 
for t in tax_meas: 
    tax_df[t+'pct'] = tax_df[t]/tax_df['rev_inc_sc']
    print(t+'pct')
    
# generate other variable for both types of countries 
tax_df['add_rc'] = tax_df['direct_inc_sc_ex_resource_revpct'] + tax_df['tot_resource_revpct'] + tax_df['tax_g_spct'] + tax_df['tax_int_trade_transpct'] 
tax_df['add_nonrc'] = tax_df['direct_inc_scpct'] + tax_df['tax_g_spct'] + tax_df['tax_int_trade_transpct'] 

tax_df['other_rc'] = 1 - tax_df['add_rc']
tax_df['other_nonrc'] = 1 - tax_df['add_nonrc']
```

### Generate countrycode and year variable by separating the indentifier variable 

```python
tax_df['countrycode']= tax_df['identifier'].astype(str).str[0:3]
tax_df['year'] = df_tax['identifier'].astype(str).str[3:7].astype(float)
```

### Merge selected variables with the main dataset

```python
# list only those that are needed
tax_vars = ['countrycode', 'year', 'tot_resource_revpct', 'direct_inc_sc_ex_resource_revpct', 'tax_g_spct', 'tax_int_trade_transpct', 'other_rc', 'other_nonrc','direct_inc_scpct']
tax_df = tax_df[tax_vars]

# Rename to match others in detail
tax_vars = ['countrycode', 'year', 'Tax: Total Resource Revenue (% of total)', 'Tax: Direct Income Revenue - Excluding Resource Income (% of total)', 'Tax: Goods and Service Tax Revenue (% of total)', 'Tax: International Trade Revenue (% of total)', 'Tax: Other Inclduing Resources (% of total)', 'Tax: Other - Excluding Resources (% of total)','Tax: Direct Revenue - Including from resouces (% of total)']
tax_df.columns = tax_vars
```

```python
### Scale Tax variables: 
for i in (list(set(tax_vars)- set(['countrycode', 'year']))): 
    tax_df[i] = tax_df[i]*100
```

### SQL Database 

The odi-data-portal works through Postgres databases within the web server.  Thus, to generate the portal locally prior to deployment, we upload the data to a local Postgres server.  

#### Define Connect Function

```python

def connect(user, password, db, host='localhost', port=5432):
    '''Returns a connection and a metadata object'''
    # We connect with the help of the PostgreSQL URL
    # postgresql://federer:grandestslam@localhost:5432/tennis
    url = 'postgresql://{}:{}@{}:{}/{}'
    url = url.format(user, password, host, port, db)

    # The return value of create_engine() is our connection object
    engine = sqlalchemy.create_engine(url, client_encoding='utf8')
    
    ## Commections 
    connection = engine.connect()
    
    # We then bind the connection to MetaData()
    meta = sqlalchemy.MetaData(bind=engine, reflect=True)

    return engine, connection, meta
```
